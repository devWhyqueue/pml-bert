```python
from datasets import load_from_disk, DatasetDict, Dataset

dataset = load_from_disk('/kaggle/working/civil_comments')['train']
toxicity_subtypes = ["severe_toxicity", "obscene", "threat", "insult", "identity_attack", "sexual_explicit"]
empty_dataset = Dataset.from_dict({col: [] for col in dataset.column_names})
for subtype in toxicity_subtypes:
    filtered_data = dataset.filter(lambda x: x[subtype] > 0.5)
    dataset_dict = DatasetDict({
        "train": empty_dataset,
        "validation": empty_dataset,
        "test": filtered_data
    })
    dataset_dict.save_to_disk(f"/kaggle/working/civil_comments/{subtype}")
```

```python
for subtype in toxicity_subtypes:
    dataset_path = f"/kaggle/working/civil_comments/{subtype}"
    !python replibert/main.py tokenize --dataset_dir "{dataset_path}" --text_field text --destination "{dataset_path}_tokenized"
```

```python
for subtype in toxicity_subtypes:
    print(f"Evaluating for subtype {subtype}...")
    dataset_path = f"/kaggle/working/civil_comments/{subtype}_tokenized"
    !torchrun --nproc_per_node=1 --rdzv_backend=c10d replibert/main.py evaluate --dataset_name civil_comments --dataset_dir "{dataset_path}" --eval_set test --weight_file "/kaggle/input/toxicbert-weights/pytorch/default/1/civil_plus_jigsaw_test_0.69.pth"
```
